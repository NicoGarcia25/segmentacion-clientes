```python
# Importar las librerías necesarias
import pandas as pd
import gdown

# Descargar el archivo desde Google Drive
# Se eligió usar gdown para evitar la necesidad de montar Google Drive o subir el archivo manualmente,
# asi se facilita la ejecución de la notebook en cualquier entorno sin requerir autenticación.
gdown.download('https://drive.google.com/uc?id=1oGTaIUhj1krV0-luYAQVs_c2IdVnvZaS', 'Mall_Customers (1).csv')

# Leer el dataset
# El arhivo es tipo csv, asi que su separador por defecto es la coma, pero se indica explícitamente con sep=',' por claridad.
df = pd.read_csv('Mall_Customers (1).csv', sep=',')

# Mostrar las primeras filas del DataFrame para verificar
print(df.head())
```

```python
# Renombrar columnas para facilitar su uso
df.rename(columns={'Annual Income (k$)': 'Ingreso_Anual', 'Spending Score (1-100)': 'Puntaje_Gasto'}, inplace=True)

# Visualización interactiva 1: Edad vs Ingreso Anual (Diferenciado por Género)
fig1 = px.scatter(df, x="Age", y="Ingreso_Anual", color="Gender",
                  title="Relación entre Edad e Ingreso Anual",
                  labels={"Age": "Edad", "Ingreso_Anual": "Ingreso Anual (k$)"},
                  hover_data=["Puntaje_Gasto"],
                  template="plotly_dark",
                  color_discrete_sequence=px.colors.qualitative.Pastel)

# Visualización interactiva 2: Ingreso Anual vs Puntaje de Gasto (Nuestra clave para el clustering)
fig2 = px.scatter(df, x="Ingreso_Anual", y="Puntaje_Gasto", color="Age",
                  title="Ingreso Anual vs. Puntaje de Gasto (Mapa de calor por Edad)",
                  labels={"Ingreso_Anual": "Ingreso Anual (k$)", "Puntaje_Gasto": "Puntaje de Gasto (1-100)"},
                  template="plotly_dark",
                  color_continuous_scale="Viridis")

fig1.show()
fig2.show()
```

## ⚙️ Armado del Modelo (Clustering Particional: K-Means)

El clustering particional divide los datos en un número predefinido de grupos y se usa ampliamente en la segmentación de clientes para categorizar audiencias según sus comportamientos. Para confirmar matemáticamente cuántos clústeres necesitamos realmente (y validar nuestra intuición visual de los 5 grupos), usaremos el **Método del Codo (Elbow Method)**.


```python
# Seleccionamos las características para el modelo (Ingreso Anual y Puntaje de Gasto)
X = df[['Ingreso_Anual', 'Puntaje_Gasto']]

# Escalado de datos (Buena práctica en clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Método del Codo para encontrar el K óptimo
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Gráfico del Método del Codo interactivo
fig_elbow = px.line(x=list(range(1, 11)), y=wcss, markers=True,
                    title='Método del Codo para determinar K óptimo',
                    labels={'x': 'Número de Clústeres (K)', 'y': 'WCSS (Inercia)'},
                    template="plotly_dark")
fig_elbow.update_traces(line_color='cyan', marker=dict(size=10, color='magenta'))
fig_elbow.show()
```

### 📝 Análisis del Codo y Entrenamiento:

La gráfica del Método del Codo evalúa la inercia (WCSS) frente a la cantidad de clústeres. Como se puede observar, la curva muestra una caída drástica y pronunciada desde K=1 hasta K=4, lo que indica que el algoritmo está logrando reducir significativamente la varianza interna al crear nuevos grupos.

El punto de inflexión clave (el "codo") se produce exactamente en **K=5**. A partir de este punto (K=6 en adelante), la disminución de la inercia se vuelve marginal y la curva se estabiliza de forma casi plana. Esto nos confirma matemáticamente lo que habíamos intuido visualmente en el análisis exploratorio: segmentar en más de 5 grupos solo sobrecomplicaría el modelo sin aportar un valor real o diferencial. Por lo tanto, entrenaremos nuestro modelo K-Means definitivo para segmentar a los clientes en estos 5 grupos operativos estratégicos.


```python
# Entrenamos el modelo definitivo con K=5
kmeans_final = KMeans(n_clusters=5, init='k-means++', random_state=42)
df['Cluster'] = kmeans_final.fit_predict(X_scaled)

# Convertir a string para visualización categórica en Plotly
df['Cluster_Label'] = df['Cluster'].astype(str)

# DICCIONARIO CORREGIDO: Mapeo exacto según los centroides de random_state=42
cluster_names = {
    '0': 'Ingreso Medio - Gasto Medio (Estándar)',
    '1': 'Ingreso Alto - Gasto Alto (Clientes Estrella)',
    '2': 'Ingreso Bajo - Gasto Alto (Impulsivos)',
    '3': 'Ingreso Alto - Gasto Bajo (Ahorradores)',
    '4': 'Ingreso Bajo - Gasto Bajo (Conservadores)'
}
df['Perfil_Cliente'] = df['Cluster_Label'].map(cluster_names)

# Mapa de colores fijo para mantener la coherencia visual en el informe
color_map = {
    'Ingreso Medio - Gasto Medio (Estándar)': 'green',
    'Ingreso Alto - Gasto Alto (Clientes Estrella)': 'purple',
    'Ingreso Alto - Gasto Bajo (Ahorradores)': 'orange',
    'Ingreso Bajo - Gasto Alto (Impulsivos)': 'blue',
    'Ingreso Bajo - Gasto Bajo (Conservadores)': 'red'
}

# Gráfico final de Clústeres
fig_clusters = px.scatter(df, x="Ingreso_Anual", y="Puntaje_Gasto", color="Perfil_Cliente",
                          title="Segmentación Final de Clientes (K-Means)",
                          labels={"Ingreso_Anual": "Ingresos Anuales (k$)", "Puntaje_Gasto": "Puntaje de Gastos"},
                          hover_data=["Age", "Gender"],
                          template="plotly_dark",
                          color_discrete_map=color_map) # Usamos el mapa forzado

# Agregar los centroides al gráfico
centroids = scaler.inverse_transform(kmeans_final.cluster_centers_)
fig_clusters.add_trace(go.Scatter(x=centroids[:, 0], y=centroids[:, 1],
                                  mode='markers', marker=dict(color='yellow', size=15, symbol='star'),
                                  name='Centroides'))

fig_clusters.show()
```
